# Use Case

When you think of professional baseball it always seems to turn to the battle of offense and defense. The game has changed a lot over the years with one being more dominant than the other from time to time. Now more than ever the game is at a very balanced state but as MLB analysts we want to know from the dataset how much has it really changed over the past 50 years. In 1968 the MLB decided to change the mound height because of how good pitchers were performing, specifically Bob Gibson’s 1.68 ERA that year. Years later the stats for batters became slightly better which showed that the MLB’s changes had improved the game to some extent. With this data we want to study how much changing the height of the mound affected the batters statistics by looking at hits, homeruns, rbi’s, walks, and a few other measures. This will be very important to know since baseball has changed a lot over the course of fifty years. It can help the MLB analyze if the changes made have really made a great impact on batting statistics. Also the MLB wants to know over the years what standout hitters were actually inducted into the Hall of Fame. Finding out this information will help point out whether it was just the batters skills or if they could have possibly used PED’s to help improve their stats.

# Graph Algorithms

# Louvain

The louvain algorithm allows a user to detect communities in large networks and maximizes a modularity score for each community. In this example team id was used in order to figure out how many communities were there with the same id. In the future this could be used to easily identify communities within another given database for the MLB

# Label Propogation

This algorithm demonstrates label propogation and goes through two iterations to help figure out communities within the pitching dataset. It takes a look at teamID and shows all the unique values for ERA in the database. This form of algorithm is important to the MLB analysts because not only does it show a unique community for each node but also from that we can see the importance of the maximum number of neighboring nodes that are not unique.
